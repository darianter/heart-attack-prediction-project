```python
import requests

url = 'http://127.0.0.1:8000/predict'
load = {'data': 'heart_test.csv'}

response = requests.post(url, json=load)

if response.status_code == 200:
    print('Предсказания получены:')
    print(response.json())
else:
    print(f'Ошибка: {response.status_code}')
    print(response.text)
```
